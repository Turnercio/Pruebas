# Visual Studio Code



[Visual Studio Code](https://code.visualstudio.com/), también conocido como VSCode, es un editor de código fuente desarrollado por Microsoft. Es un software gratuito y de código abierto que está disponible para Windows, macOS y Linux. Visual Studio Code es altamente personalizable y extensible, lo que permite a los desarrolladores adaptarlo a sus necesidades y preferencias.

Visual Studio Code es el _IDE_ más popular del momento, con diferencia y el que usaremos en este y otros muchos módulos.

## Instalación

Ve a [su web](https://code.visualstudio.com/) y sigue las instrucciones de instalación para tu sistema operativo.

## Extensiones

Una de las ventajas de vscode, son la extensiones.on complementos creados por la comunidad que añaden funcionalidad adicional al editor. Estas extensiones pueden abarcar desde linters y formateadores de código hasta integraciones con servicios en la nube y herramientas de colaboración en tiempo real.

De momento sólo instalaremos unas pocas, para cambiar el aspecto del editor. Hay muchas más, que necesitarás si empiezas a usar otros lenguajes o herramientas como git. De momento, no toqueteamos más.

### Instalar extensiones

Selecciona extensiones an la barra lateral y escribe:

1. "Grueling hare"
1. "darcula"

Instala ambas extensiones.

### CMD + SHIFT + P : Paleta de comandos

Un atajo de teclado de que debes de aprender es el que te muestra la paleta de comandos. Este te desplegará un menú con comandos que puedes aplicar.

Prueba y escribe `theme`. Te saldrán varias opciones de autocompletar. Selecciona `Preferences : Color Theme`.  Elije algún esquema de color y fuentes que te guste, por ejemplo alguno de os que acabas de instalar.

### Ejercicio

Selecciona las extensiones en la barra lateral e instala las siguientes:

1. **Back & Forth**: Te añade unas flechas para moverte por los ficheros, similares a las felchas en un navegador. Las tendrás en la barra superior a la derecha.
2. **Git Graph**: La usarás cuando aprendas `git` un poco más adelante.
3. **Jupyter Notebook Renderers**: permite ver los notebooks de Jupyter dentro del propio VSCode.
